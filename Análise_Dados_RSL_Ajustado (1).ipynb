```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
import re
from collections import Counter

sns.set(style="whitegrid")

# Baixar stopwords (se necessário)
nltk.download("stopwords")
from nltk.corpus import stopwords

# --------------------------------------------------
# 1. Carregamento dos Dados e Merge
# --------------------------------------------------
file_path_1 = "/content/drive/MyDrive/Documentos Mestrado/Materiais e Documentos /Documentos Grupo de Pesquisa /Artigo Científico REDUÇÃO DE DIMENSIONALIDADE DE DADOS/Artigo RSL/articles (2).xls"
file_path_2 = "/content/drive/MyDrive/Documentos Mestrado/Materiais e Documentos /Documentos Grupo de Pesquisa /Artigo Científico REDUÇÃO DE DIMENSIONALIDADE DE DADOS/Artigo RSL/data_extraction-_1_-_1_.xlsx"

articles_data = pd.read_excel(file_path_1)
data_extraction = pd.read_excel(file_path_2)

articles_data["title"] = articles_data["title"].astype(str)
data_extraction["article"] = data_extraction["article"].astype(str)

if "year" in articles_data.columns:
    articles_data["Ano"] = articles_data["year"]

print("Valores únicos na coluna 'Ano' em articles_data:")
print(articles_data["Ano"].unique())

data_combined = pd.merge(articles_data, data_extraction, left_on="title", right_on="article", how="inner")
print("\nValores únicos na coluna 'Ano' após o merge:")
print(data_combined["Ano"].unique())

# --------------------------------------------------
# 2. Definir Palavras-Chave e Funções de Extração
# --------------------------------------------------

# Lista de palavras-chave que indicam nomes de algoritmos relevantes
core_keywords = [
    "pca", "t-sne", "tsne", "umap", "ica", "k-means", "mds", "svd",
    "autoencoder", "cnn", "louvain", "clustering", "regressão logística",
    "random forest", "spectral", "isomap", "netboost", "firefly", "lsm",
    "sammon"
]

def unify_expressions(text):
    """
    Unifica variações de escrita para algoritmos.
    Exemplo: 'tsne', 't sne', 't.sne' -> 't-sne'; 'kmeans', 'k means' -> 'k-means'
    """
    text = text.lower()
    # PCA
    text = re.sub(r"\bp\.?\s?c\.?\s?a\b", "pca", text)
    text = re.sub(r"\bpca\b", "pca", text)
    # t-SNE
    text = re.sub(r"\btsne\b", "t-sne", text)
    text = re.sub(r"\bt\.?\s?-\s?sne\b", "t-sne", text)
    text = re.sub(r"\bt\.?\s?sne\b", "t-sne", text)
    # k-means
    text = re.sub(r"\bkmeans\b", "k-means", text)
    text = re.sub(r"\bk\s?-\s?means\b", "k-means", text)
    text = re.sub(r"\bk\s?means\b", "k-means", text)
    # SVD
    text = re.sub(r"\bs\.?\s?v\.?\s?d\b", "svd", text)
    return text

def extract_algorithm_names(text, max_words=10):
    """
    1. Unifica expressões.
    2. Divide o texto usando delimitadores: nova linha, vírgula, ponto-e-vírgula.
    3. Limpa cada candidato e mantém apenas os que tenham no máximo 'max_words'
       e que contenham pelo menos uma palavra-chave de core_keywords.
    """
    text = unify_expressions(text)
    # Dividir por nova linha, vírgula ou ponto-e-vírgula
    candidates = re.split(r"[\n,;]+", text)
    extracted = []
    for cand in candidates:
        cand = cand.strip(" -•\r\t.").strip()
        if not cand:
            continue
        words = cand.split()
        if len(words) > max_words:
            continue  # descarta frases muito longas (possivelmente comentários)
        # Verifica se contém pelo menos uma core keyword
        if any(kw in cand for kw in core_keywords):
            extracted.append(cand)
    return extracted

# --------------------------------------------------
# 3. Extração de Frases de "Algoritmos Utilizados"
# --------------------------------------------------
data_combined["Extracted_Names"] = data_combined["Algoritmos Utilizados"].apply(lambda x: extract_algorithm_names(x, max_words=10))
extracted_exploded = data_combined.explode("Extracted_Names")

# Contar a frequência das frases extraídas
name_counter = Counter(extracted_exploded["Extracted_Names"].tolist())
print("\nAlgoritmos extraídos (frases) mais frequentes:")
for name, freq in name_counter.most_common(15):
    print(f"{name}: {freq}")

# --------------------------------------------------
# 4. Visualização dos Resultados
# --------------------------------------------------
import pandas as pd
top_names = pd.DataFrame(name_counter.most_common(10), columns=["Algoritmo", "Frequência"])
plt.figure(figsize=(10, 5))
plt.bar(top_names["Algoritmo"], top_names["Frequência"], color="lightcoral", width=0.8)
plt.title("Top 10 Algoritmos Extraídos de 'Algoritmos Utilizados'")
plt.xlabel("Algoritmo")
plt.ylabel("Frequência")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
import re
from collections import Counter
from nltk.util import ngrams
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

sns.set(style="whitegrid")

# Baixar stopwords se necessário
nltk.download("stopwords")
from nltk.corpus import stopwords

# =============================================================================
# SEÇÃO 1: Carregamento dos Dados e Merge
# =============================================================================
# Caminhos conforme informados (não foram alterados)
file_path_1 = "/content/drive/MyDrive/Documentos Mestrado/Materiais e Documentos /Documentos Grupo de Pesquisa /Artigo Científico REDUÇÃO DE DIMENSIONALIDADE DE DADOS/Artigo RSL/articles (2).xls"
file_path_2 = "/content/drive/MyDrive/Documentos Mestrado/Materiais e Documentos /Documentos Grupo de Pesquisa /Artigo Científico REDUÇÃO DE DIMENSIONALIDADE DE DADOS/Artigo RSL/data_extraction-_1_-_1_.xlsx"

# Carregar os datasets
articles_data = pd.read_excel(file_path_1)
data_extraction = pd.read_excel(file_path_2)

# Converter para string as colunas usadas para merge
articles_data["title"] = articles_data["title"].astype(str)
data_extraction["article"] = data_extraction["article"].astype(str)

# Criar a coluna 'Ano' a partir de 'year', se existir
if "year" in articles_data.columns:
    articles_data["Ano"] = articles_data["year"]

print("Valores únicos na coluna 'Ano' em articles_data:")
print(articles_data["Ano"].unique())

# Merge usando "title" (articles_data) e "article" (data_extraction)
data_combined = pd.merge(
    articles_data,
    data_extraction,
    left_on="title",
    right_on="article",
    how="inner"
)

print("\nValores únicos na coluna 'Ano' após o merge:")
print(data_combined["Ano"].unique())

# =============================================================================
# SEÇÃO 2: Extração de Algoritmos (Expressões/Frases) de "Algoritmos Utilizados"
# =============================================================================
# Lista de palavras-chave que indicam nomes de algoritmos relevantes
core_keywords = [
    "pca", "t-sne", "umap", "ica", "k-means", "mds", "svd",
    "autoencoder", "cnn", "lsm", "spectral", "isomap", "regressão logística",
    "random forest", "netboost", "firefly", "sammon"
]

def unify_expressions(text):
    """
    Unifica variações comuns para algoritmos:
      - 'tsne', 't sne', 't.sne' -> 't-sne'
      - 'kmeans', 'k means', 'k-means' -> 'k-means'
      - 'p c a', 'pca' -> 'pca'
      - 's v d' -> 'svd'
    """
    text = text.lower()
    text = re.sub(r"\bp\.?\s?c\.?\s?a\b", "pca", text)
    text = re.sub(r"\bpca\b", "pca", text)
    text = re.sub(r"\btsne\b", "t-sne", text)
    text = re.sub(r"\bt\.?\s?-\s?sne\b", "t-sne", text)
    text = re.sub(r"\bt\.?\s?sne\b", "t-sne", text)
    text = re.sub(r"\bkmeans\b", "k-means", text)
    text = re.sub(r"\bk\s?-\s?means\b", "k-means", text)
    text = re.sub(r"\bk\s?means\b", "k-means", text)
    text = re.sub(r"\bs\.?\s?v\.?\s?d\b", "svd", text)
    return text

def split_all_phrases(text):
    """
    Divide o texto em frases utilizando delimitadores comuns: nova linha, ponto-e-vírgula e vírgula.
    """
    text = unify_expressions(text)
    phrases = re.split(r"[\n;,]+", text)
    phrases = [phrase.strip(" -•\r\t.") for phrase in phrases if phrase.strip()]
    return phrases

def extract_candidates(text, max_words=10):
    """
    Divide o texto em candidatos usando delimitadores e mantém somente aqueles com até max_words
    que contenham pelo menos uma core keyword.
    """
    candidates = split_all_phrases(text)
    filtered = []
    for cand in candidates:
        cand_clean = cand.strip()
        words = cand_clean.split()
        if len(words) > max_words:
            continue
        if any(kw in cand_clean.lower() for kw in core_keywords):
            filtered.append(cand_clean)
    return filtered

# Extrair expressões de "Algoritmos Utilizados"
data_combined["Extracted_Names"] = data_combined["Algoritmos Utilizados"].apply(lambda x: extract_candidates(x, max_words=10))
extracted_exploded = data_combined.explode("Extracted_Names")

# Contar a frequência das expressões extraídas
name_counter = Counter(extracted_exploded["Extracted_Names"].tolist())
print("\nAlgoritmos extraídos (frases) mais frequentes:")
for name, freq in name_counter.most_common(15):
    print(f"{name}: {freq}")

# =============================================================================
# SEÇÃO 3: Tendência dos Algoritmos ao Longo dos Anos
# =============================================================================
def split_algorithms(algs):
    return [x.strip() for x in str(algs).split(",") if x.strip()]

df_alg = data_combined.copy()
df_alg["Algoritmos_Lista_Simple"] = df_alg["Algoritmos Utilizados"].apply(split_algorithms)
df_exploded_simple = df_alg.explode("Algoritmos_Lista_Simple")
counts_by_year = df_exploded_simple.groupby(["Ano", "Algoritmos_Lista_Simple"]).size().reset_index(name="Frequência")
print("\nTabela de frequências dos algoritmos por ano:")
print(counts_by_year.head(10))

top_algorithms = counts_by_year["Algoritmos_Lista_Simple"].value_counts().head(5).index
trend_data = counts_by_year[counts_by_year["Algoritmos_Lista_Simple"].isin(top_algorithms)]
plt.figure(figsize=(10,6))
for alg in top_algorithms:
    subset = trend_data[trend_data["Algoritmos_Lista_Simple"] == alg]
    plt.plot(subset["Ano"], subset["Frequência"], marker="o", label=alg)
plt.title("Tendência dos Algoritmos Mais Populares ao Longo dos Anos")
plt.xlabel("Ano")
plt.ylabel("Frequência")
plt.legend()
plt.grid(True)
plt.show()

# =============================================================================
# SEÇÃO 4: Cluster de Subáreas (Áreas da Biomedicina)
# =============================================================================
if "Área da Biomedicina Abordada" in data_extraction.columns:
    areas = data_extraction["Área da Biomedicina Abordada"].dropna().tolist()
    vectorizer = TfidfVectorizer(stop_words="english")
    X = vectorizer.fit_transform(areas)
    kmeans = KMeans(n_clusters=5, random_state=42)
    clusters = kmeans.fit_predict(X)
    df_areas = pd.DataFrame({"Área": areas, "Cluster": clusters})
    print("\nExemplos de clusters de áreas:")
    for c in range(5):
        print(f"\nCluster {c}:")
        subset = df_areas[df_areas["Cluster"] == c]["Área"].tolist()
        for area in subset[:5]:
            print(" -", area)
    pca = PCA(n_components=2, random_state=42)
    X_reduced = pca.fit_transform(X.toarray())
    plt.figure(figsize=(8,6))
    sns.scatterplot(x=X_reduced[:,0], y=X_reduced[:,1], hue=clusters, palette="viridis")
    plt.title("Visualização dos Clusters de Áreas (PCA)")
    plt.xlabel("PC1")
    plt.ylabel("PC2")
    plt.legend(title="Cluster")
    plt.show()
else:
    print("A coluna 'Área da Biomedicina Abordada' não foi encontrada no dataset de extração.")

# =============================================================================
# SEÇÃO 5: Associação entre Algoritmos de Redução e Técnicas Complementares
# =============================================================================
DR_algorithms = {"pca", "t-sne", "umap", "ica", "svd", "mds"}
complementary_techniques = {"random forest", "svm", "k-means", "cnn", "logistic regression", "autoencoder"}

df_exploded_simple["Algoritmo_Simple"] = df_exploded_simple["Algoritmos_Lista_Simple"].apply(lambda x: x.lower())
grouped = df_exploded_simple.groupby("article").agg({
    "Algoritmo_Simple": lambda x: set([a for a in x if a in DR_algorithms]),
    "Algoritmos_Lista_Simple": lambda x: set([a.lower() for a in x if a.lower() in complementary_techniques])
}).reset_index()

associations = []
for idx, row in grouped.iterrows():
    for dr in row["Algoritmo_Simple"]:
        for comp in row["Algoritmos_Lista_Simple"]:
            associations.append((dr, comp))
df_assoc = pd.DataFrame(associations, columns=["DR", "Complementar"])
assoc_counts = df_assoc.value_counts().reset_index(name="Frequência")
print("\nAssociação entre algoritmos de redução (DR) e técnicas complementares:")
print(assoc_counts)

# =============================================================================
# SEÇÃO 6: Análise de Áreas Mais Abordadas nos Artigos
# =============================================================================
if "Área da Biomedicina Abordada" in data_extraction.columns:
    area_counts = data_extraction["Área da Biomedicina Abordada"].value_counts().reset_index()
    area_counts.columns = ["Área", "Frequência"]
    print("\nTop 10 Áreas Mais Abordadas:")
    print(area_counts.head(10))

    plt.figure(figsize=(8,5))
    plt.bar(area_counts["Área"].head(5), area_counts["Frequência"].head(5), color="green")
    plt.title("Top 5 Áreas Mais Abordadas")
    plt.xlabel("Área")
    plt.ylabel("Frequência")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()
else:
    print("A coluna 'Área da Biomedicina Abordada' não foi encontrada no dataset de extração.")
```
